### Data preprocessing

```python
import numpy as np
import pandas as pd
```

```python
movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head(1)
```

```python
credits.head(1)
```

```python
movies=movies.merge(credits,on='title')
print(movies.shape)
```

```python
print(movies.columns)
```

```python
#genres
#id
#keywords
#title
#overview
#cast
#crew
movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
movies.head(1)
```

The final dataframe shall consist of 3 columns: movie_id, title, overview. In the overview column, we will add the genres, keywords, cast(top 3) and crew columns.

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0]['genres']#list of dictionaries in a string
```

```python
import ast
```

```python
def convert(object):#helper function
    L=[]
    for i in ast.literal_eval(object):
        L.append(i['name'])
    return L
```

```python
convert(movies.iloc[0]['genres'])
```

```python
movies['genres']=movies['genres'].apply(convert)
movies.iloc[0]['genres']
```

```python
movies['keywords'][3]
```

```python
movies['keywords']=movies['keywords'].apply(convert)
print(movies.iloc[3]['keywords'])
```

```python
#for cast we'll just extract the top 3 casts
def convert3(object):
    L=[]
    for i in range(3):
        try:
            L.append(ast.literal_eval(object)[i]['name'])
        except:
            pass
    return L
print(convert3(movies.iloc[0]['cast']))
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies['cast'][40]
```

```python
movies.head(3)
```

```python
ast.literal_eval(movies['crew'][0])
```

```python
#now we'll extract the names of the directors
def fetch_director(object):
    L=[]
    for i in ast.literal_eval(object):
        if i['job']=='Director':
            L.append(i['name'])
    return L
movies['crew']=movies['crew'].apply(fetch_director)
movies.head(1)
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
print(movies.iloc[3]['overview'])
```

```python
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
print(movies.iloc[3]['cast'])
```

```python
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
print(movies.iloc[3]['crew'])
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
df=movies[['tags','title', 'movie_id']]
df.head(3)
```

```python
df['tags']=df['tags'].apply(lambda x:" ".join(x))
df.head(3)
```

### Vectorization

```python
df['tags'][0]
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)

df['tags']=df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
vectors=cv.fit_transform(df['tags']).toarray()
```

```python
print(vectors.shape, len(df))
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)#distance of each movie with every other movie
print(similarity.shape)
```

```python
similarity[0][:5]
```

### Main function

```python
df.columns
```

```python
df.head(5)
```

```python
df[df['title']=='Avatar'].index[0]
```

```python
def recommend(movie):
    movie_index=df[df['title']==movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(df.iloc[i[0]].title)
```

```python
recommend('The Dark Knight Rises')
```
